```python
def display(s,ilist):
    
    print("###########################Group{} ##############################################".format(s))
    for i in range(len(ilist)):
        img = cv2.imread(ilist[i],1)
        img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
        plt.subplot(4,3,i+1)
        plt.imshow(img)
        plt.xticks([])
        plt.yticks([])

    plt.show()
```
